```
import pandas as pd
#reading the csv using pandas read_csv function
df = pd.read_csv("classification_5000.csv")

#display the file read
df
```

```
#import the function to randomly split dataset into training and testing
from sklearn.model_selection import train_test_split

#extracts the quality type to create a y data frame
y = df["Quality_type"]

#drop the unnecessary columns and add the rest to X dataframe
X = df.drop(["DateTime","Quality_type"],axis=1)

#split X and y vectors into training (80%) and testing (20%)
X_train,X_test,y_train,y_test = train_test_split(X,y,test_size=0.20,random_state=22)

print("X_train =", X_train.shape)
print("X_test = ", X_test.shape)
print("y_train = ", y_train.shape)
print("y_test = ", y_test.shape)
```

```
#importing machine learning algorithm : random forest regressor
from sklearn.ensemble import RandomForestClassifier

#declare rfr as random forest algorithm
rfr = RandomForestClassifier()

#launch random forest training
rfr.fit(X_train,y_train)
```

```
#evalutes the performance of the training set
rfr.score(X_train,y_train)
```

```
#evalute performance of the test set 
rfr.score(X_test,y_test)
```

```
#comparing prediction with actual values
df["Quality_prediction"] = rfr.predict(X)
df.sample(15, random_state=22)
```

```
#raw features importance display
rfr.feature_importances_
```

```
#creating dataframe with the features importance
rfr_imp = pd.DataFrame(rfr.feature_importances_,columns=["Importance"])

#multiplying the importance value with 100 to ease the reading
rfr_imp["Importance"] = rfr_imp["Importance"]*100

#using the features names as index
rfr_imp = rfr_imp.set_index([X.columns])

#displayingt the dataframe with desecending importance values
display(rfr_imp.sort_values(by="Importance", ascending=False))

```
